However, this changes the style globally for the rest of the session.

You can use a context manager to set the style temporarily, which we’ll do for the rest of this article.

```python
with plt.style.context('classic'):
  plt.plot()
```

We plot a simple line graph in the code below using all the default styles.

You can also view all the stylesheets in the official reference.

https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html

```python
# source: https://www.dunderdata.com/blog/view-all-available-matplotlib-styles

x = np.arange(-2, 8, .1)
y = .1 * x ** 3 - x ** 2 + 3 * x + 2

fig = plt.figure(dpi=100, figsize=(10, 20), tight_layout=True)
available = ['default'] + plt.style.available
for i, style in enumerate(available):
    with plt.style.context(style):
        ax = fig.add_subplot(10, 3, i + 1)
        ax.plot(x, y)
    ax.set_title(style)
```

Now let’s check out the Rosé Pine theme.

We define a custom function that plots a 2x2 grid of subplots of a bar plot, a scatter plot, a line plot, and a box plot.

This way, you have an idea of what the theme looks like for different charts.

```python
def create_subplots():
    """
    Creates a 2x2 grid of subplots with different types of plots.
    """
    np.random.seed(0)
    fig, ax = plt.subplots(2, 2, figsize=(11, 8))
    
    # Bar plot
    x = np.arange(10)
    y = np.random.rand(10)
    ax[0, 0].bar(x, y)

    # Scatter plot
    x = np.random.rand(100)
    y = np.random.rand(100)
    ax[0, 1].scatter(x, y, s=100)

    # Line plot
    ax[1, 0].plot(np.random.randn(30))

    # Boxplot
    data = [np.random.normal(0, std, size=100) for std in range(1, 4)]
    labels = ['x1', 'x2', 'x3']
    bplot1 = ax[1, 1].boxplot(data,
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks

    colors = ['pink', 'lightblue', 'lightgreen']
    for patch, color in zip(bplot1['boxes'], colors):
        patch.set_facecolor(color)
```

Here’s the ggplot theme.

You’ll find it familiar if you’re an R user (more specifically, a ggplot2 fan)

```python
with plt.style.context('ggplot'):
    create_subplots()
```

Here’s another great theme by the popular FiveThirtyEight blog.

```python
with plt.style.context('fivethirtyeight'):
  create_subplots()
```

This is vanilla Rosé Pine.

If you want it slightly brighter, use the Rosé Pine Moon theme.

```python
with plt.style.context("/tmp/rose-pine.mplstyle"):
  create_subplots()
```

If you want to switch to light mode, use Rosé Pine Dawn.

```python
with plt.style.context("/tmp/rose-pine-dawn.mplstyle"):
  create_subplots()
```
